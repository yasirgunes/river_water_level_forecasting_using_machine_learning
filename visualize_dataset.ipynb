```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('dataset.csv')

# Convert the 'datetime' column to datetime objects
df['datetime'] = pd.to_datetime(df['datetime'])

# Create the plot
plt.figure(figsize=(15, 7))
plt.plot(df['datetime'], df['stage_m'])

# Add titles and labels for clarity
plt.title('Stage Height Over Time')
plt.xlabel('Date')
plt.ylabel('Stage Height (ft)')
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Set datetime as index
df.set_index('datetime', inplace=True)

# Check for gaps
df['missing'] = df['stage_m'].isna()
missing_by_year = df['missing'].resample('YE').mean()

# Plot missing data by year
plt.figure(figsize=(12, 6))
missing_by_year.plot.bar()
plt.title('Percentage of Missing Data by Year')
plt.xlabel('Year')
plt.ylabel('Percentage of Missing Data')
plt.grid(True)
plt.show()


# Display basic statistics
print("\nBasic Statistics:")
print(df[['stage_m']].describe())
```
